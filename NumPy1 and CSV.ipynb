```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,11)

# random data
np.random.seed(10)
y=np.random.randint(1,100,10)
plt.plot(x,y,marker='*',color='r',markersize=10,mfc='black')
plt.show()
```

```python
plt.figure(figsize=(13,6))
t=np.arange(0,2,0.01)
s=np.sin(2*np.pi*t)
h=np.cos(2*np.pi*t)
plt.subplot(2,1,1)
plt.plot(t,s,label='sin')
plt.legend()
plt.subplot(2,1,2)
plt.plot(t,h,label='cos')
plt.legend()
plt.show()
```

# WORKING ON CSV FILES


```python
import matplotlib.pyplot as plt
import csv
x=[]
y=[]
with open('Book1.csv','r') as f:
    data=csv.reader(f)
    for col in data:
        x.append(col[0])
        y.append(col[1])
print('x= ',x)
print('y= ',y)
plt.plot(x,y)
plt.show()
```
