# Data cleaning

```python
import pandas as pd
data=pd.read_csv('nba.csv')
data
#dropna command
data.dropna()
```

```python
data.dropna(how='any') #If any NA values are present, drop that row or column.
```

```python
data.dropna(how='all') # If all values are NA, drop that row or column.
```

```python
data.dropna(subset=['Salary','College'])#Searches the salary
```

# fillna Command

```python
data.fillna(0)
data['Salary'].fillna(0)
data
```

```python
data['College'].fillna('No College',inplace=True)
data
```

```python
#sort_values method with data frame
data.sort_values('Name',ascending=False)
data.sort_values('Salary',na_position='first')
```

```python
data.sort_values(['Team','Salary'],ascending=[False,True])
```

```python
data['Salary']=data['Salary'].fillna(0).astype('int')
data
```

```python
data['Salary Rank']=data['Salary'].rank(ascending=False).astype('int')
data
data.sort_values('Salary',ascending=False)
```

```python
import pandas as pd
data=pd.read_csv('employees.csv')
mask1=data['Team']=='Legal'
mask2=data['Team']=='Sales'
mask3=data['Team']=='Product'
data[mask1 | mask2 | mask3]
# Shortcut for the above operation
# ❤️isin command
mask=data['Team'].isin(['Legal','Sales','Product'])
data[mask]
data[data['Team'].isin(['Legal','Sales','Product'])]
```
